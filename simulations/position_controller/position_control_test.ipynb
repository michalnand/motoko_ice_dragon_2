```python
import numpy
import matplotlib.pyplot as plt


data = numpy.loadtxt("data/position_test_1.txt")

print(data.shape)

sample_start = 0
sample_stop = -1

time    = data[sample_start:sample_stop, 0]

distance_req    = data[sample_start:sample_stop, 1]*1000.0
angle_req       = data[sample_start:sample_stop, 2]*180.0/numpy.pi

distance            = data[sample_start:sample_stop, 3]*1000.0
velocity            = data[sample_start:sample_stop, 4]*1000.0
angle               = data[sample_start:sample_stop, 5]*180.0/numpy.pi
angular_velocity    = data[sample_start:sample_stop, 6]*180.0/numpy.pi


plt.clf()
plt.figure(figsize=(11, 8))
plt.plot(time, distance_req,   color="red", label="required distance", lw=3.0)
plt.plot(time, distance, color="blue", label="obtained distance", alpha=0.5)
#plt.plot(time, 0.02*velocity, color="green", label="velocity", alpha=0.5)
plt.xlabel("time [ms]")
plt.ylabel("distance [mm]")
plt.legend()
plt.show()



plt.clf()
plt.figure(figsize=(11, 8))
plt.plot(time, angle_req,   color="red", label="required angle", lw=3.0)
plt.plot(time, angle, color="blue", label="angle", alpha=0.5)
#plt.plot(time, 0.1*angular_velocity, color="green", label="angular velocity", alpha=0.5)
plt.xlabel("time [ms]")
plt.ylabel("angle [degrees]")
plt.legend()
plt.show()
```
