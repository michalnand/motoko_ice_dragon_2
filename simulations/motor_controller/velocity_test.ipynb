```python
import numpy
import matplotlib.pyplot as plt


data = numpy.loadtxt("velocity_0.txt")

print(data.shape)


time    = data[:, 0]
ref     = data[:, 1]

result_raw   = data[:, 2]   
result_fil   = data[:, 3]

#result_raw   = data[:, 4]
#result_fil   = data[:, 5]


plt.clf()
plt.figure(figsize=(11, 8))

plt.plot(time, ref,   color="red", label="required velocity", lw=3.0)
plt.plot(time, result_raw, color="blue", label="raw reading", alpha=0.5)
plt.plot(time, result_fil, color="green", label="filtered", alpha=0.5)

plt.xlabel("time [ms]")
plt.ylabel("velocity [rpm]")
plt.legend()
plt.show()

'''
t_start = 1200

time    = data[t_start:, 0]
ref     = 1000*data[t_start:, 1]
result   = data[t_start:, 2]
left    = data[t_start:, 3]

plt.clf()
plt.figure(figsize=(11, 8))

plt.plot(time, ref,   color="red", label="torque", lw=3.0)
plt.plot(time, result, color="blue", label="result wheel", alpha=0.5)
plt.plot(time, left, color="green", label="left wheel", alpha=0.5)

plt.xlabel("time [ms]")
plt.ylabel("velocity [rpm]")
plt.legend()
plt.show()  
'''
```
