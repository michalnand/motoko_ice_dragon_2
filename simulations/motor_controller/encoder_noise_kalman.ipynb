```python
import numpy
import matplotlib.pyplot as plt


data = numpy.loadtxt("encoder_noise_3.txt")

print(data.shape)


time    = data[:, 0]
ref     = 1000*data[:, 1]

result_raw        = data[:, 2]
result_fil        = data[:, 3]

#result_raw        = data[:, 4]
#result_fil        = data[:, 5]


plt.clf()
plt.figure(figsize=(11, 8))

plt.plot(time, ref,       color="red",   label="torque", lw=3.0)
plt.plot(time, result_raw, color="green", label="raw reading", alpha=0.5)
plt.plot(time, result_fil, color="blue",  label="filtered", alpha=0.8)

plt.xlabel("time [ms]")
plt.ylabel("velocity [rpm]")
plt.legend()
plt.show()
```
